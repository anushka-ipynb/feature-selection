```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('housingprice.csv')
```

```python
df.head(10)
```

#### main aim is to find the sales price. we're only using the numerical features in this case, because we'll have to handle categorical features, which we don't have to for this explanation.

```python
df.info()
```

```python
house_df = df.select_dtypes(include=np.number)
```

```python
house_df.drop(['Id'],axis=1,inplace=True)
```

```python
X = house_df.drop(['SalePrice'],axis=1)
y = house_df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size=0.3)
```

```python
X_train.isnull().sum()
```

```python
from sklearn.feature_selection import mutual_info_regression
mutual_info = mutual_info_regression(X_train.fillna(0),y_train)
```

```python
mutual_info = pd.Series(mutual_info)
mutual_info.index = X_train.columns
```

```python
mutual_info.sort_values(ascending=False)
```

```python
from sklearn.feature_selection import SelectPercentile
```

```python
selected_top_columns = SelectPercentile(mutual_info_regression, percentile=20)
selected_top_columns.fit(X_train.fillna(0), y_train)
X_train.columns[selected_top_columns.get_support()]
```

### these are the columns that belong to the top 20th percentile
